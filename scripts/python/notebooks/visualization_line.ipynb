```python
import pandas as pd

from bokeh.io import output_notebook, show
from bokeh.plotting import figure
```

```python
output_notebook()
```

```python
DATA_FILE = 'Experiments/Experiment1000/data/1001_fNIRS_All_Data_s1.csv'
COND_FILE = 'Experiments/Experiment1000/data/1001_fNIRS_conditions_s1.csv'
```

```python
data = pd.read_csv(DATA_FILE)
cond = pd.read_csv(COND_FILE)
```

```python
start = int(cond['Task 1'][0])
end = start + int(cond['Task 1'][1])
```

```python
y = [data['CH1'][i] for i in range(start, end)]
x = range(start, end)
```

```python
p = figure(plot_width=900, plot_height=300, title="My Line Plot")
p.line(x, y)
show(p)
```
